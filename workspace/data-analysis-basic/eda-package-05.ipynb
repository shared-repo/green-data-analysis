```python
# 구분 컬럼을 제외한 나머지 컬럼의 데이터를 수치형으로 변경
columns = seoul_cctv.columns
print( columns[1] )
for column in columns[1:]: # 총계, 2013이전, ...
    seoul_cctv[column] = seoul_cctv[column].map(lambda v: int(v.replace(",", "")))

seoul_cctv.info()
```

```python
# 컬럼이름 변경
seoul_cctv.rename(columns={"구분":"자치구"}, inplace=True)
seoul_cctv.head()
```
